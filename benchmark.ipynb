### 3.1 – Bubble Sort

Bubble Sort is a simple algorithm which is used to sort a given set of n elements provided in form of an array with n number of elements. Bubble Sort compares all the element one by one and sort them based on their values.

If the given array has to be sorted in ascending order, then bubble sort will start by comparing the first element of the array with the second element, if the first element is greater than the second element, it will swap both the elements, and then move on to compare the second and the third element, and so on.

If we have total n elements, then we need to repeat this process for n-1 times.

It is known as bubble sort, because with every complete iteration the largest element in the given array, bubbles up towards the last place or the highest index, just like a water bubble rises up to the water surface.

Sorting takes place by stepping through all the elements one-by-one and comparing it with the adjacent element and swapping them if required.

#### 3.1.1 – Time Complexity

To calculate the complexity of the bubble sort algorithm, it is useful to determine how many comparisons each loop performs. For each element in the array, bubble sort does n−1 comparisons. In Big O notation, bubble sort performs O(n) comparisons. 

Because the array contains n elements, it has an O(n) numbers of elements, so the running time is O(n^2).

#### 3.1.2 – Bubble Code

```python
# Author : Alexander Pepe
# Bubble Sort Algorithm

# Code adapted from https://www.geeksforgeeks.org/bubble-sort/

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# function's name
def bubble_sort(arr):
    # reassign the range of length as i
    for i in range(len(arr)): 
        # find the numbers of outer passes
        for j in range(len(arr) - 1): 
            # the condition works if the first element is greater than the next
            if arr[j] > arr[j+1]: 
                # swap the elements if follow the condition
                arr[j], arr[j+1] = arr[j+1], arr[j] 
    # return the array sorted
    return arr
```

### 3.2 – Insertion Sort

It is another sorting methods in which the number of comparisons depends on the order of the input elements. It works on the principle of a sorted item with one item at a time. An item is taken from the array, it is inserted at its correct position by comparing the element for its neighbour. The process is repeated until there is no more unsorted item in the list.

![title](https://www.w3resource.com/w3r_images/insertion-sort.png)

#### 3.2.1 – Time Complexity

The Worst-Case Complexity(O(n^2)) occurs when the array is unsorted. Each element has to be compared with each of the other elements so, for every n element, (n-1) number of comparisons are made. The total number of comparisons is n*(n-1) ~ n^2).

The Best-Case Complexity(O(n)) occurs when the array is already sorted, the out loop runs for n number of times whereas the inner loop does not run at all. So, there are only n number of comparisons. The complexity is called linear.

The Average-Case Complexity(O(n2)) occurs when the elements of an array are in jumbled order (neither ascending nor descending).

#### 3.2.2 – Insertion Code

```python
# Author : Alexander Pepe
# Insertion Sort Algorithm

# Code adapted from https://www.geeksforgeeks.org/insertion-sort/

# function's name
def insertion_sort(arr):
    # reassign the elements in a range between 1 and array's length as i
    for i in range(1, len(arr)):
        # storing current element
        key = arr[i]
        # check whether the adjacent element in left side
        #  is greater or less than the current element
        j = i
        while j > 0 and arr[j - 1] > key :
                arr[j] = arr[j - 1]
                j -= 1
        arr[j] = key
    return arr
```

### 3.3 – Heap Sort

Heapsort is an in-place, comparison-based sorting algorithm and can be thought of as an improved selection sort as it divides the input into a sorted and an unsorted region. It iteratively shrinks the unsorted region by extracting the largest/smallest element and moving that to the sorted region. The improvement consists of using a heap data structure rather than a linear-time search to find the maximum.
Heapsort does not produce a stable sort, which means that the implementation does not preserve the input order of equal elements in the sorted output.

The heapsort algorithm can be divided into two parts:

    o	In the first step, a heap is built out of the input data. We can do this in O(n) time.
    
    o	In the second step, a sorted array is created by repeatedly removing the largest/smallest element from the heap (the root of the heap) and inserting it into the array. The heap is updated (heapify-down is called on the root node) after each removal to maintain the heap property. Once all elements have been removed from the heap, we get a sorted array. This is done in O(n.log(n)) time since we need to pop n elements, where each removal involves a constant amount of work and a single heapify operation takes O(log(n)) time.

![title](https://cdn.programiz.com/cdn/farfuture/VicaT2DyDXxbtM88OYklajepD4hkdSumEHTg2nBwe7s/mtime:1586942728/sites/tutorial2program/files/heap_sort.png)

#### 3.3.1 – Heap Code

```python
# Author : Alexander Pepe
# Heap Sort Algorithm

# code adapted from https://docs.python.org/3/library/heapq.html

from heapq import heappop, heappush # import those libraries to run the code

def heap_sort(arr):
    heap = []
    for elem in arr:
        heappush(heap, elem) # push value item onto the heap

    ordered = []

    # While we have elements left in the heap
    while heap:
        ordered.append(heappop(heap)) # pop and return the smallest item from the heap

    return ordered
```

### 3.4 – Bucket Sort

This algorithm works dividing the elements into several groups called Bucket. The elements inside groups are sorted using any of the suitable sorting algorithms or recursively calling the same algorithm.

This algorithm uses Insertion Sort to sort the elements in the bucket.

![title](https://static.javatpoint.com/tutorial/daa/images/daa-bucket-sort-step2.png)

#### 3.4.1 – Time Complexity

The Worst-Case Complexity(O(n^2)), elements of close range are placed in the same bucket, so some buckets having more elements the others.

The complexity depends on the sorting algorithm used to sort the elements of the bucket.

The Best-Case Complexity(O(n+k)) occurs when the elements are uniformly distributed in the buckets and when the elements inside buckets are already sorted. O(n) is the complexity for making the buckets and O(k) is the complexity for sorting the elements of the bucket using algorithms having linear time complexity at the best case.

The Average Case Complexity(O(n)) occurs when the elements are distributed randomly. I hold true until the sum of the squares of the bucket sizes is linear in the total number of elements.

#### 3.4.2 – Bucket Sort Code

```python
# Author : Alexander Pepe
# Bucket Sort Algorithm

# Code adapted from https://stackabuse.com/bucket-sort-in-python/

# function's name
def bucket_sort(arr):
    max_value = max(arr) # the maximum key value in the array
    size_value = max_value/len(arr)
    
    bucket_list = [] # new array of n empty lists
    for x in range(len(arr)):
        bucket_list.append([]) # Populating the bucket with the list elements
        
    for i in range(len(arr)): # put the array in correct bucket
        j = int(arr[i]/ size_value)
        if j != len(arr):
            bucket_list[j].append(arr[i])
        else:
            bucket_list[len(arr) - 1].append(arr[i])
            
    for z in range(len(arr)):
        insertion_sort(bucket_list[z])
        
    final_output = []
    for i in range(len(arr)):
        final_output += bucket_list[x]
    return final_output
```

### 3.5 – Counting Sort

Counting sort is a sorting algorithm that sorts the elements of an array by counting the number of occurrences of each unique element in the array. The count is stored in an auxiliary array and the sorting is done by mapping the count as an index of the auxiliary array. The figure below describes easily.

![title](https://static.wixstatic.com/media/db1234_9425d1913f5f4b0884da06e66cfef6a0~mv2.jpg/v1/fill/w_1000,h_839,al_c,q_90/db1234_9425d1913f5f4b0884da06e66cfef6a0~mv2.webp)

#### 3.5.1 - Time Complexity

For scanning the input array elements, the loop iterates n times, thus taking O(n) running time. The sorted array B[] also gets computed in n iterations, thus requiring O(n) running time. The count array also uses k iterations, thus has a running time of O(k). Thus, the total running time for counting sort algorithm is O(n+k).

#### 3.5.2 – Counting Sort Code

```python
# Author : Alexander Pepe
# Counting Sort Algorithm

# code adapted form https://www.geeksforgeeks.org/counting-sort/

# function's name
def counting_sort(arr):
    max_element = int(max(arr))
    min_element = int(min(arr))
    range_of_elements = max_element - min_element + 1
    # create an array to store individual elements
    count_arr = [0 for _ in range(range_of_elements)]
    output_arr = [0 for _ in range(len(arr))]
 
    # store each element
    for i in range(0, len(arr)):
        count_arr[arr[i]-min_element] += 1
 
    # actual positions will be created
    for i in range(1, len(count_arr)):
        count_arr[i] += count_arr[i-1]
 
    # create the array of outputs
    for i in range(len(arr)-1, -1, -1):
        output_arr[count_arr[arr[i] - min_element] - 1] = arr[i]
        count_arr[arr[i] - min_element] -= 1
 
    # new array with sorted elements
    for i in range(0, len(arr)):
        arr[i] = output_arr[i]
 
    return arr
```

### 3.6 – Shell Sorting

Shell sort is an optimization over insertion sort. This is achieved by repeatedly doing insertion sort on all elements at fixed, decreasing intervals. Last iteration the interval is 1. The intervals are known as gap sequence. Here it becomes a regular insertion sort, and it guarantees that the array will be sorted. But to note one point is that by the time we do that array is almost sorted, hence the iteration is very fast.

![title](https://miro.medium.com/max/1400/1*0t7dxb4otndZQFJW387Wtw.png )

### 3.6.1 - Time Complexity

The time complexity depends on the gap sequence. Its best-case time complexity is O (n* log n) and worst case is O (n* log 2n). Time complexity of Shell sort is generally assumed to be near to O(n) and less than O(n2) as determining its time complexity is still an open problem.

### 3.6.2 – Shell Sort Code

```python
# Author : Alexander Pepe
# Shell Sort Algorithm

# code adapt from https://www.geeksforgeeks.org/shellsort/

def shell_sort(arr, x):
 
    h = x // 2 # reduce the gap
    while h > 0: # addition elements until the entire array is sorted
        for i in range(h, x):
            t = arr[i]
            # put the gap-sorted elements in correct location
            j = i
            while j >= h and arr[j - h] > t:
                arr[j] = arr[j - h]
                j -= h
            # Now put "t" in correct position
            arr[j] = t
        h = h // 2
```

```python
# use the time library to measure the running time
shellSort = time_time2(shell_sort)

# the mean of each input size
shellSort100 = np.mean(shellSort[:10])
shellSort250 = np.mean(shellSort[10:20])
shellSort500 = np.mean(shellSort[20:30])
shellSort750 = np.mean(shellSort[30:40])
shellSort1000 = np.mean(shellSort[40:50])
shellSort1250 = np.mean(shellSort[50:60])
shellSort2500 = np.mean(shellSort[60:70])
shellSort3750 = np.mean(shellSort[70:80])
shellSort5000 = np.mean(shellSort[80:90])
shellSort6250 = np.mean(shellSort[90:100])
shellSort7500 = np.mean(shellSort[100:110])
shellSort8750 = np.mean(shellSort[110:120])
shellSort10000 = np.mean(shellSort[120:])

# Use Pandas to create DataFrame with the values by input size
dataShell = [
    ['100', shellSort100], ['250', shellSort250], 
    ['500', shellSort500], ['750', shellSort750], 
    ['1000', shellSort1000], ['1250', shellSort1250],
    ['2500', shellSort2500], ['3750', shellSort3750],
    ['5000', shellSort5000], ['6250', shellSort6250],
    ['7500', shellSort7500], ['8750', shellSort8750],
    ['10000', shellSort10000]
    ]

dfShell = pd.DataFrame(dataShell, columns=['Size', 'Time'])
# lambda function plus round to print three decimal place
dfShell["Time"] = dfShell["Time"].apply(lambda x: round(x, 3))

dfShell
```

### 3.7 - Pancake Sorting

From Wikipedia:

    - Pancake sorting is the colloquial term for the mathematical problem of sorting a disordered stack of pancakes in order of size when a spatula can be inserted at any point in the stack and used to flip all pancakes above it.
    
The goal is to order the elements from largest on the bottom, to smallest on the top. The pancake sorting asks how many such "prefix reversals" are sufficient to sort an arbitrary array.

It follows the idea of flip the stacks. The Fig. 12 illustrated it.

![title](https://iq.opengenus.org/content/images/2020/07/pancake_Sort.png)

#### 3.7.1 - Time Complexity

The number of flips required to sort the array is in the order of 'n'. In the best case, the array is sorted, and no flips are required. 
The worst-case occurs when the elements are alternating small and large elements. A total of 2n-3 flips is required. The number of flips required is of the order of O(n). The worst-case time complexity is [Big O]: O(n2).

The best-case occurs when the array is already sorted, and no flips are required. The best-case time complexity is [Big Omega]: O(n).

#### 3.7.2 – Pancake Sort Code

```python
# Author : Alexander Pepe
# Pancake Sort Algorithm

# code adapted from https://www.geeksforgeeks.org/pancake-sorting/

i = len(size)
# flip until the lagest is on bottom 
def flip(arr, i): 
    start = 0
    while start < i: 
        temp = arr[start] 
        arr[start] = arr[i] 
        arr[i] = temp 
        start += 1
        i -= 1


def findMax(arr, n): # index of the maximum
    mi = 0
    for i in range(0,n): 
        if arr[i] > arr[mi]: 
            mi = i 
    return mi 
  
# sorting using flip 
def pancake_sort(arr, n): 
    
    # reduce the size by one
    curr_size = n 
    while curr_size > 1: 
        # Find index of the maximum element  
        mi = findMax(arr, curr_size) 
  
        # move the maximum element to the end
        if mi != curr_size-1: 
            flip(arr, mi) 
  
            # move the maximum number by reversing array
            flip(arr, curr_size-1) 
        curr_size -= 1
    return arr
```

```python
# use the time library to measure the running time
pancakeSort = time_time2(pancake_sort)

# the mean of each input size
pancakeSort100 = np.mean(pancakeSort[:10])
pancakeSort250 = np.mean(pancakeSort[10:20])
pancakeSort500 = np.mean(pancakeSort[20:30])
pancakeSort750 = np.mean(pancakeSort[30:40])
pancakeSort1000 = np.mean(pancakeSort[40:50])
pancakeSort1250 = np.mean(pancakeSort[50:60])
pancakeSort2500 = np.mean(pancakeSort[60:70])
pancakeSort3750 = np.mean(pancakeSort[70:80])
pancakeSort5000 = np.mean(pancakeSort[80:90])
pancakeSort6250 = np.mean(pancakeSort[90:100])
pancakeSort7500 = np.mean(pancakeSort[100:110])
pancakeSort8750 = np.mean(pancakeSort[110:120])
pancakeSort10000 = np.mean(pancakeSort[120:])

# Use Pandas to create DataFrame with the values by input size
dataPancake = [
    ['100', pancakeSort100], ['250', pancakeSort250], 
    ['500', pancakeSort500], ['750', pancakeSort750], 
    ['1000', pancakeSort1000], ['1250', pancakeSort1250],
    ['2500', pancakeSort2500], ['3750', pancakeSort3750],
    ['5000', pancakeSort5000], ['6250', pancakeSort6250],
    ['7500', pancakeSort7500], ['8750', pancakeSort8750],
    ['10000', pancakeSort10000]
    ]

dfPancake = pd.DataFrame(dataPancake, columns=['Size', 'Time'])
# lambda function plus round to print three decimal place
dfPancake["Time"] = dfPancake["Time"].apply(lambda x: round(x, 3))

dfPancake
```
